```python
import pandas as pd
df=pd.read_csv('sortingText.csv')

print(df)
```

```python
alldf = df.iloc[:, 2:]
print(alldf)
```

```python
alldf.sum(axis=0).sort_values(ascending=False)
```

```python
import nltk
nltk.download('wordnet')

from nltk.corpus import wordnet as wn
```

```python
ss = wn.synsets('color', pos = wn.NOUN)[0]
hyps = list(set(
                [w for s in ss.closure(lambda s:s.hyponyms())
                        for w in s.lemma_names()]))
sorted(hyps)
```

```python
colorlist= [i for i in list(alldf) if i in hyps]
print(colorlist)
```

```python
colorsdf=alldf[colorlist]
colorsdf.sum(axis=0).sort_values(ascending=False)
```

```python
repurchasedf = df[df['Target'] == 1 ]
colorlist1= [i for i in list(repurchasedf) if i in hyps]
colorsdf1=repurchasedf[colorlist1]
colorsdf1.sum(axis=0).sort_values(ascending=False)
```

```python
ss2 = wn.synsets('material', pos = wn.NOUN)[2]
hyps2 = list(set(
                [w for s in ss2.closure(lambda s:s.hyponyms())
                        for w in s.lemma_names()]))
sorted(hyps2)
```

```python
materiallist = [i for i in alldf if i in hyps2]
materialdf=alldf[materiallist]
materialdf.sum(axis=0).sort_values(ascending=False)
```

```python
materiallist1= [i for i in list(repurchasedf) if i in hyps2]
materialdf1=repurchasedf[materiallist1]
materialdf1.sum(axis=0).sort_values(ascending=False)
```
